```python
%pylab inline
import numpy as np
from scipy.fftpack import fft, fftfreq
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
#Cargando la voz
#http://nbviewer.ipython.org/github/mgeier/python-audio/blob/master/audio-files/audio-files-with-scipy-io.ipynb
fs, sig = wavfile.read('manuel_hw5.wav')
n=len(sig) #para futura normalizacion de la transf. de Fourier
plt.plot(sig) #Grafica de la voz
```

```python
#http://nbviewer.ipython.org/github/forero/ComputationalMethods/blob/master/notes/fourier.ipynb
fft_rand_signal = fft(sig) / n # FFT Normalized
freq = fftfreq(n, 1) # Recuperamos las frecuencias
plt.plot(freq,np.abs(fft_rand_signal))
#grafica de la transformada de fourier de la voz
```

```python
#Sacando el espectro de potencias
half_n = np.ceil(n) 
fft_rand_signal_half = (2.0) * fft_rand_signal[:half_n]
freq_half = freq[:half_n]
#El arreglo del Freq_half es de una dimension, mientras que el de fft_rand_signal_half es dos dimensiones. No se pueden graficar
#por tener distintas dimensiones, por lo que voy a sacar un promedio del fft_rand_signal_half, para volverlo de una dimension
scatter(freq_half, log10(np.absolute((fft_rand_signal_half[:,0]+fft_rand_signal_half[:,1])/2)**2)) 
```

### Ruido Blanco

```python
blancofourier=fft_rand_signal_half/np.absolute(fft_rand_signal_half)
plot(np.abs(blancofourier))
plt.title('Ruido Blanco Fourier')

```

```python
#Invirtiendo el ruido blanco, y tomando los valores reales
blanco =ifft(blancofourier) 
plot(blanco.real)
plt.title('Ruido blanco ')


```

```python
#Se guarda este archivo en el repositorio como:
wavfile.write('ruidoblanco.wav', fs, blanco.real)
```

### Ruido rosa

```python
#El ruido rosa es el ruido blanco en fourier sobre la frecuencia (1/freq)

rosafourier = ((blancofourier[:,0])+(blancofourier[:,0])/2)/(freq) #como el ruido blanco es un arreglo de dos dimensiones, saque el promedio de las
#dos dimensiones, para hacerlo de una dimension, y asi poderlo dividir por el arreglo unidimensional de la frecuencia. 
rosafourier1 = (blancofourier[:,0])/(freq)
rosafourier2 = (blancofourier[:,0])/(freq)
#Grafica del ruido rosa

#La gráfica no se comporta como 1/frecuencia porque debe tener un valor muy pequeño que hace que tenga esa especie de asintota,
# si se acerca la gráfica, se observa que si tiene un comportamiento como deberia
plt.subplot(141)
plot((np.abs(rosafourier)))
plt.title('Ruido rosa ')

plt.subplot(142)
plot((rosafourier))
plt.title('Ruido rosa promedio de columnas')#Grafica si se saca un promedio del arreglo del ruido Blanco en Fourier
plt.xlim(0,100)
plt.ylim(0, 10000)

plt.subplot(143)
plot((rosafourier1))              #Grafica si sólo se tiene en cuenta la primer columna del arreglo del ruido Blanco en Fourier
plt.title('Ruido rosa primer columna')
plt.xlim(0,100)
plt.ylim(0, 10000)

plt.subplot(144)
plot((rosafourier2))              #Grafica si sólo se tiene en cuenta la segunda columna del arreglo del ruido Blanco en Fourier
plt.title('Ruido rosa segunda columna')
plt.xlim(0,100)
plt.ylim(0, 10000)

# Se puede ver que las graficas tienen un comportamiento similar 
```

```python
#Invirtiendo el ruido rosa de fourier
rosa = ifft(rosafourier) #No se observa transformada inversa de fourier porque, como se puede observar en la primer imagen anterior
#la gráfica no converge a cero, sino que diverge, y transformadas inversas de Fourier sólo se presentan cuando hay 
#convergencia a cero. Por lo tanto, creo que es la razon de que no se observa una gráfica de la transf inversa.
plot(rosa)
plt.title('Ruido rosa en Fourier')
```
